```python
data=pd.read_csv('/Users/watchara.p/Desktop/Gene_UPDEG_ML.csv',low_memory=False)
data.head()
```

```python
x=data.iloc[:,1 :171]
y=data.iloc[:,0]

x
y
```

```python
from sklearn.preprocessing import StandardScaler ,LabelEncoder


scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

```python
numbers = range(len(x_scaled))
groups = [f"Group {num // 3 + 1}" for num in numbers]
groups
```

```python
from sklearn.metrics import confusion_matrix, precision_score, f1_score
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import numpy as np

# Assuming x_scaled and y are your features and labels, and `groups` is defined for LOGO
kf = LeaveOneGroupOut()

model_rf = RandomForestClassifier(random_state=99)
model_sv = SVC(kernel='linear', probability=True, random_state=99)
model_lr = LogisticRegression(random_state=99)
model_gbc = GradientBoostingClassifier(random_state=99)

# Lists to store aggregated true and predicted labels for all models
y_true_all = []
y_pred_rf_all = []
y_pred_sv_all = []
y_pred_lr_all = []
y_pred_gbc_all = []


# Performance metrics lists
acc_rf, acc_sv, acc_gbc, acc_lr = [], [], [], []
pre_rf, pre_sv, pre_gbc, pre_lr = [], [], [], []
sens_rf, sens_sv, sens_gbc, sens_lr = [], [], [], []
spec_rf, spec_sv, spec_gbc, spec_lr = [], [], [], []
f11_rf, f11_sv, f11_gbc, f11_lr = [], [], [], []
```

```python
# LOGO loop
for train_index, test_index in kf.split(x_scaled, groups=groups):
    X_train, X_test = x_scaled.iloc[train_index], x_scaled.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit the models
    model_gbc.fit(X_train, y_train)
    model_sv.fit(X_train, y_train)
    model_rf.fit(X_train, y_train)
    model_lr.fit(X_train, y_train)

    # Predict
    y_pred_lr = model_lr.predict(X_test)
    y_pred_sv = model_sv.predict(X_test)
    y_pred_rf = model_rf.predict(X_test)
    y_pred_gbc = model_gbc.predict(X_test)

    # Store true and predicted labels for aggregation
    y_true_all.extend(y_test)  # Add true labels
    y_pred_lr_all.extend(y_pred_lr)
    y_pred_sv_all.extend(y_pred_sv)
    y_pred_rf_all.extend(y_pred_rf)
    y_pred_gbc_all.extend(y_pred_gbc)


# Aggregated confusion matrix and metrics
y_true_all = np.array(y_true_all)
y_pred_rf_all = np.array(y_pred_rf_all)
y_pred_sv_all = np.array(y_pred_sv_all)
y_pred_lr_all = np.array(y_pred_lr_all)
y_pred_gbc_all = np.array(y_pred_gbc_all)

cm_rf_total = confusion_matrix(y_true_all, y_pred_rf_all, labels=labels)
cm_sv_total = confusion_matrix(y_true_all, y_pred_sv_all, labels=labels)
cm_lr_total = confusion_matrix(y_true_all, y_pred_lr_all, labels=labels)
cm_gbc_total = confusion_matrix(y_true_all, y_pred_gbc_all, labels=labels)

print("Aggregated Confusion Matrix (RF):\n", cm_rf_total)
print("Aggregated Confusion Matrix (GBC):\n", cm_gbc_total)
print("Aggregated Confusion Matrix (LR):\n", cm_lr_total)
print("Aggregated Confusion Matrix (SVM):\n", cm_sv_total)
```
