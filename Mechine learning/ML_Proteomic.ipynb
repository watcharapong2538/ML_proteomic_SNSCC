```python
import os
import numpy as np  
import matplotlib.pyplot as mtp  
import pandas as pd  
```

# Load and Prepare Data

```python
data=pd.read_csv('/Users/watchara.p/Desktop/Gene_UPDEG_ML.csv',low_memory=False)
data.head()
```

# Assuming the first column is the label and the rest are features
# x = Features (proteomic data) (cloumn 2-171)
# y = Labels (SNSCC=1 and NP=0) (first column)

```python
x=data.iloc[:,1 :171]
y=data.iloc[:,0]

x
y
```

# Normalize and scale the data

```python
from sklearn.preprocessing import StandardScaler ,LabelEncoder


scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

# Split data into training and testing sets
# train_test_split: Splits the data into training (80%) and test_size=0.2 (testing (20%) sets), maintaining the same class distribution in both sets (stratification).
# stratify make the y have the same proportion between Y train and Y test

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import LabelEncoder
from matplotlib import pyplot as plt

X_train, X_test, Y_train, Y_test = train_test_split(x_scaled, y, test_size=0.2, stratify=y, random_state=2)



print ('X_test number', X_test.shape[0])
print ('X_train number', X_train.shape[0])
count1 = np.count_nonzero(Y_train)
count2 = np.count_nonzero(Y_test)
print("Count of SNSCC in Y_train:", count1)
print("Count of SNSCC in Y_test:", count2)
```

# Train and Evaluate Models

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, KFold, LeaveOneOut
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, precision_recall_curve, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import time
import statistics
```

# Model Training and Evaluation: Each model is trained on the training data and evaluated using cross-validation (n_splits=5) (KFold and LeaveOneOut) and various performance metrics (accuracy, precision, recall, F1-score, AUROC, AUPRC, specificity, and sensitivity).
# use random_state = 0 for  


```python
algo = [
    [RandomForestClassifier(random_state=0), 'RandomForest'],
    [SVC(kernel='linear', probability=True, random_state=0), 'Support vector machine'],
    [LogisticRegression(random_state=0), 'LogisticRegression'],
    [GradientBoostingClassifier(random_state=0), 'Gradientboost']
]

model_scores = []
result = []
looc = LeaveOneOut()
kf = KFold(n_splits=5)
```

```python

for a in algo:
    start_time = time.time()
    algo_name = a[1]
    model = a[0]
    model.fit(X_train, Y_train)

    # Cross-validation scores
    cvkfold = cross_val_score(model, X_train, Y_train, cv=kf)
    cvlooc = cross_val_score(model, X_train, Y_train, cv=looc)

    # Model predictions and evaluation (Performance)
    y_pred = model.predict(X_test)
    ac = accuracy_score(Y_test, y_pred)  # Accuracy score
    r = model.score(X_test, Y_test)
    model_scores.append([r, algo_name])
    precision1 = precision_score(Y_test, y_pred)  # Precision
    recall1 = recall_score(Y_test, y_pred)  # Recall
    f1 = f1_score(Y_test, y_pred)  # F1-Score
    fpr, tp, thresholds = roc_curve(Y_test, model.predict_proba(X_test)[:, 1])
    AUROC = auc(fpr, tp)  # Area Under ROC Curve
    precision, recall, thresholds = precision_recall_curve(Y_test, model.predict_proba(X_test)[:, 1])
    AUPRC = auc(recall, precision)  # Area Under Precision-Recall Curve


    confusion = confusion_matrix(Y_test, y_pred)
    total1 = sum(sum(confusion))
    accuracy1 = (confusion[0, 0] + confusion[1, 1]) / total1  # Accuracy from confusion matrix
    specificity1 = confusion[0, 0] / (confusion[0, 0] + confusion[0, 1])  # Specificity
    sensitivity1 = confusion[1, 1] / (confusion[1, 0] + confusion[1, 1])  # Sensitivity (Recall)
    SD_KF = statistics.stdev(cvkfold)  # Standard deviation of KFold CV scores
    SD_loocv = statistics.stdev(cvlooc)  # Standard deviation of LeaveOneOut CV scores

    # Compile results
    values = [cvkfold.mean(), cvlooc.mean(), ac, r, precision1, recall1, f1, AUROC, AUPRC, specificity1, sensitivity1, SD_KF, SD_loocv]
    result.append(values)

    print(f'Model: {algo_name}, time = {(time.time() - start_time)} sec')
    if a[1] in ['RandomForest','Gradientboost']:
        features = pd.Series(model.feature_importances_,index=x_scaled.columns).sort_values(ascending=False)
        ft=pd.DataFrame(features)
        ft.to_csv('/Users/watchara.p/Desktop/'+a[1]+'_features.csv',index=True)
    elif a[1] == 'LogisticRegression' :
        features = pd.Series(model.coef_[0],index=x_scaled.columns).sort_values(ascending=False)
        ft=pd.DataFrame(features)
        ft.to_csv('/Users/watchara.p/Desktop/'+a[1]+'_features.csv',index=True)
    else :
        support_vectors = model.support_vectors_
        dual_coefficients = model.coef_        
        ft = pd.DataFrame(dual_coefficients.T,index=x_scaled.columns)
        ft2 = ft.sort_values(by=ft.columns[0],ascending=False)
        ft2.to_csv('/Users/watchara.p/Desktop/'+a[1]+'_features.csv',index=True)



print(f'best importance score = {max(model_scores)}')    
```
