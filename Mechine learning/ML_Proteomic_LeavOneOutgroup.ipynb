# Load and Prepare Data

```python
data=pd.read_csv('/Users/watchara.p/Desktop/Gene_UPDEG_ML.csv',low_memory=False)
data.head()
```

# Assuming the first column is the label and the rest are features
# x = Features (proteomic data) (cloumn 2-171)
# y = Labels (SNSCC=1 and NP=0) (first column)

```python
x=data.iloc[:,1 :171]
y=data.iloc[:,0]

x
y
```

# Normalize and scale the data

```python
from sklearn.preprocessing import StandardScaler ,LabelEncoder


scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

# Divides samples into sequential groups of 3 (e.g., Group 1: samples 0â€“2, Group 2: samples 3â€“5, etc.).

```python
numbers = range(len(x_scaled))
```

```python
groups = [f"Group {num // 3 + 1}" for num in numbers]
groups
```

# Perform train-test split based on groups
# train_test_split: Splits the data into training (80%) and test_size=0.2 (testing (20%) sets).

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GroupShuffleSplit
from matplotlib import pyplot as plt

# Perform train-test split based on groups
group_splitter = GroupShuffleSplit(test_size=0.2, random_state=2)
train_idx, test_idx = next(group_splitter.split(x_scaled, y, groups))

# Split data
X_train, X_test = x_scaled.iloc[train_idx], x_scaled.iloc[test_idx]
Y_train, Y_test = y.iloc[train_idx], y.iloc[test_idx]


print ('X_test number', X_test.shape[0])
print ('X_train number', X_train.shape[0])
count1 = np.count_nonzero(Y_train)
count2 = np.count_nonzero(Y_test)
print("Count of SNSCC in Y_train:", count1)
print("Count of SNSCC in Y_test:", count2)
```

```python
list(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, KFold, LeaveOneOut, LeavePOut, LeaveOneGroupOut
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_curve, auc, precision_recall_curve, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
import time
import statistics
```

# Model Training and Evaluation: Each model is trained on the training data and evaluated with various performance metrics (accuracy, precision, recall, F1-score, AUROC, AUPRC, specificity, and sensitivity).
# cross-validation (KFold (n_splits=5)) or (LeaveOneGroupOut) was performed in all dataset
# use random_state = 99 for consistency result  


```python
algo = [
    [RandomForestClassifier(random_state=99), 'RandomForest'],
    [SVC(kernel='linear', probability=True, random_state=99), 'Support vector machine'],
    [LogisticRegression(random_state=99), 'LogisticRegression'],
    [GradientBoostingClassifier(random_state=99), 'Gradientboost']
]

model_scores = []
result = []
y_pre=[]
y_te=[]
looc = LeaveOneGroupOut()
kf = KFold(n_splits=5)
```

```python
# Test LeaveOneGroupOut (LOGO)
cvlooc = cross_val_score(RandomForestClassifier(random_state=0), x_scaled, y, cv=looc,groups=groups,verbose=2)
```

```python
cvlooc.mean()
```

```python
for a in algo:
    start_time = time.time()
    algo_name = a[1]
    model = a[0]
    

    # Cross-validation scores
    #The LeaveOneGroupOut (LOGO) strategy ensured that all samples in one group were used as the test set, while the remaining groups were used for training.
    cvkfold = cross_val_score(model, x_scaled, y, cv=kf)
    cvlooc = cross_val_score(model, x_scaled, y, cv=looc,groups=groups)
    
    
    model.fit(X_train, Y_train)

    # Model predictions and evaluation (Performance)
    y_pred = model.predict(X_test)
    ac = accuracy_score(Y_test, y_pred)  # Accuracy score
    r = model.score(X_test, Y_test)
    model_scores.append([r, algo_name])
    precision1 = precision_score(Y_test, y_pred)  # Precision
    recall1 = recall_score(Y_test, y_pred)  # Recall
    f1 = f1_score(Y_test, y_pred)  # F1-Score
    fpr, tp, thresholds = roc_curve(Y_test, model.predict_proba(X_test)[:, 1])
    AUROC = auc(fpr, tp)  # Area Under ROC Curve
    precision, recall, thresholds = precision_recall_curve(Y_test, model.predict_proba(X_test)[:, 1])
    AUPRC = auc(recall, precision)  # Area Under Precision-Recall Curve



    confusion = confusion_matrix(Y_test, y_pred)
    total1 = sum(sum(confusion))
    accuracy1 = (confusion[0, 0] + confusion[1, 1]) / total1  # Accuracy from confusion matrix
    specificity1 = confusion[0, 0] / (confusion[0, 0] + confusion[0, 1])  # Specificity
    sensitivity1 = confusion[1, 1] / (confusion[1, 0] + confusion[1, 1])  # Sensitivity (Recall)
    SD_KF = statistics.stdev(cvkfold)  # Standard deviation of KFold CV scores
    SD_loocv = statistics.stdev(cvlooc)  # Standard deviation of LeaveOneOut CV scores


    ci_ac_length = 1.96 * np.sqrt((accuracy1 * (1 - accuracy1)) / Y_test.shape[0])
    ci_lower = accuracy1 - ci_ac_length
    ci_upper = accuracy1 + ci_ac_length
    ac_length = str(ci_upper) +'-'+str(ci_lower)

    ci_sense_length = 1.96 * np.sqrt((sensitivity1 * (1 - sensitivity1)) / Y_test.shape[0])
    sense_ci_lower = sensitivity1 - ci_sense_length
    sense_ci_upper = sensitivity1 + ci_sense_length
    sense_length = str(sense_ci_upper) +'-'+str(sense_ci_lower)

    ci_specifi_length = 1.96 * np.sqrt((specificity1 * (1 - specificity1)) / Y_test.shape[0])
    specifi_ci_lower = specificity1 - ci_specifi_length
    specifi_ci_upper = specificity1 + ci_specifi_length
    specifi_length = str(specifi_ci_upper) +'-'+str(specifi_ci_lower)


    ci_preci_length = 1.96 * np.sqrt((precision1 * (1 - precision1)) / Y_test.shape[0])
    preci_ci_lower = precision1 - ci_preci_length
    preci_ci_upper = precision1 + ci_preci_length
    preci_length = str(preci_ci_upper) +'-'+str(preci_ci_upper)


    # Compile results
    values = [cvkfold.mean(), cvlooc.mean(), ac, r, precision1, recall1, f1, AUROC, AUPRC, specificity1, sensitivity1, SD_KF, SD_loocv, ac_length, sense_length, specifi_length , preci_length]
    result.append(values)
    y_pre.append(y_pred)
    y_te.append(Y_test)

    print(f'Model: {algo_name}, time = {(time.time() - start_time)} sec')
    if a[1] in ['RandomForest','Gradientboost']:
        features = pd.Series(model.feature_importances_,index=x_scaled.columns).sort_values(ascending=False)
        ft=pd.DataFrame(features)
        ft.to_csv('/Users/watchara.p/Desktop/'+a[1]+'_features.csv',index=True)
    elif a[1] == 'LogisticRegression' :
        features = pd.Series(model.coef_[0],index=x_scaled.columns).sort_values(ascending=False)
        ft=pd.DataFrame(features)
        ft.to_csv('/Users/watchara.p/Desktop/'+a[1]+'_features.csv',index=True)
    else :
        support_vectors = model.support_vectors_
        dual_coefficients = model.coef_        
        ft = pd.DataFrame(dual_coefficients.T,index=x_scaled.columns)
        ft2 = ft.sort_values(by=ft.columns[0],ascending=False)
        ft2.to_csv('/Users/watchara.p/Desktop/'+a[1]+'_features.csv',index=True)



print(f'best importance score = {max(model_scores)}')    
```

```python
# Display all results
for i, (model, RandomForest) in enumerate(algo):
    print(f'Results for {RandomForest}:')
    print(f'5-fold CV mean: {result[i][0]}')
    print(f'LOOCV mean: {result[i][1]}')
    print(f'Accuracy: {result[i][2]}')
    print(f'R-squared: {result[i][3]}')
    print(f'Precision: {result[i][4]}')
    print(f'Recall: {result[i][5]}')
    print(f'F1-score: {result[i][6]}')
    print(f'AUROC: {result[i][7]}')
    print(f'AUPRC: {result[i][8]}')
    print(f'Specificity: {result[i][9]}')
    print(f'Sensitivity: {result[i][10]}')
    print(f'SD 5-fold CV: {result[i][11]}')
    print(f'SD LOOCV: {result[i][12]}')
    print(f'Accuracy 95%Cl: {result[i][13]}')
    print(f'Sensitivity 95%Cl: {result[i][14]}')
    print(f'Specificity 95%Cl: {result[i][15]}')
    print(f'Precision 95%Cl: {result[i][16]}')
    print(f'y_predict: {y_pre[i]}')
    print(f'y_test: {y_te[i]}')
    print('')
```
